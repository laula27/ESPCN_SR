## This is a README about the difference of the three folder. Details about these three folder can be seen in each folder.

In the folder for ESPCNprior work, it is implementing all the testing for different model. This is the prior work implemtation for this project.

Detailed readme will included in the folder to indicate how to run the program.

In the folder ESPCNV1, it is implementing on Set 5 dataset and T91, plwase run the prepare.py first  in order to customize the dataset and the details about how to obtain dataset and how to run prepare.py can be find in prepare.py 

After running prepare.py, please run train.py and test.py to test image or test_video.py to test video.

The detail about how to run and the where to obtain the dataset will be specified in each program.(There will be text at the top of each program to explain the datset and the step)

In the folder ESPCNV2, it us test on different dataset to try on different activate function. There is mo need to downoad dataset because it will download in the program after running the code.

Three version of ESPCN is used for different attempt, the ESPCNprior work is for testing on different model and evaluate. ESPCNV1 is for testing using customize datasets which I try using upscaling factor of threee to implement.
ESPCNV2 is for testimg on different activate function such as relu and tanh function.
